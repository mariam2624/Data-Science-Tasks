# Converting visitdate, hours and time into timestamp 

```python
df.customer_id.duplicated()
```

```python
df.dtypes
```

```python
df
```

```python
df['time'] = (pd.to_datetime(df['hitshour'].astype(str) + ':' + df['hitsminute'].astype(str), format='%H:%M').dt.time)
```

```python
df
```

```python
df['visitdate'] = pd.to_datetime(df['visitdate'])
df['Date'] = df['visitdate'].dt.strftime('%Y-%m-%d')
df['Time'] = df['visitdate'].dt.strftime('%H:%M')
```

```python
df
```

```python
df['timestamp'] = df['Date'].astype(str) + ' ' + df['time'].astype(str)
```

```python
df
```

```python
df.pop("hitshour")
df.pop("hitsminute")
df.pop("time")
df.pop("Date")
df.pop("Time")
```
