## Sorting Customer IDs

```python
df_sorted_ids = df.sort_values(by=["customer_id"])
df_sorted_ids
```

## Concatenating Customer IDs and Session IDs 

```python
df_sorted_ids['New_ID'] = df_sorted_ids.customer_id.astype(str) + df_sorted_ids.session_id.astype(str)
df_sorted_ids
```

## Converting visit date, hours and minutes into timestamp 

```python
df_sorted_ids['Time'] = (pd.to_datetime(df_sorted_ids['hitshour'].astype(str) + ':' + df_sorted_ids['hitsminute'].astype(str), format='%H:%M').dt.time)
df_sorted_ids
```

```python
df_sorted_ids['visitdate'] = pd.to_datetime(df_sorted_ids['visitdate'])
df_sorted_ids['Date'] = df_sorted_ids['visitdate'].dt.strftime('%Y-%m-%d')
#df_sorted_ids['Time'] = df_sorted_ids['visitdate'].dt.strftime('%H:%M')
```

```python
df_sorted_ids
```

```python
df_sorted_ids['Timestamp'] = df_sorted_ids['Date'].astype(str) + ' ' + df_sorted_ids['Time'].astype(str)
```
